```python
# retrieve data for 1 user

failed_users_dict = {}
keys = []
for x in fail.columns:
    if x != 'blog':
        keys.append(x)

for k in keys:
    failed_users_dict[k] = []
failed_users_dict
failed_users_dict
```

```python
for name in failed_list:
    try: 
        user = api.get_user(screen_name = name)  
        failed_users_dict['id'].append(user.id)
        failed_users_dict['fullname'].append(user.name)
        failed_users_dict['date_joined'].append(user.created_at)
        failed_users_dict['location'].append(user.location)
        failed_users_dict['num_tweets'].append(user.statuses_count)
        failed_users_dict['following'].append(user.friends_count)
        failed_users_dict['followers'].append(user.followers_count)
        failed_users_dict['likes'].append(user.favourites_count)
        failed_users_dict['lists'].append(user.listed_count)
        failed_users_dict['screen_name'].append(name)
    except:
        continue
new_fail = pd.DataFrame.from_dict(failed_users_dict)
```

```python
import numpy as np
def change_to_timestamp(s):
    if s == -1:
        return s
    new_s = s.split()
    new_s = ''.join(new_s)
    f = "%I:%M%p-%d%b%Y"
    out = datetime.strptime(new_s, f)
    out = pd.Timestamp(out)
    return out
```
